# Sistemas lineales 
# invariantes en el tiempo (LTI)


Supongamos que tenemos un sistema para el cual el estado inicial $x_0(t_0)$ y la entrada $u_0(t)$, 
$t \geq t_0$ producen la salida $y_0(t)$, $t \geq t_0$.

<img src="chenDefTmeInv.PNG"/>

Un __sistema invariante en el tiempo__ cumple para cualquier lapso de tiempo $T$ que el estado inicial $x_0(t_0 + T)$ y la entrada $u_0(t - T), t \geq t_0+T$ producen la salida $y_0(t - T), t \geq t_0+T$



Si un sistema es invariante se tiene la siguiente propiedad.
$$g(t,\tau)=g(t+T,\tau+T)=g(t-\tau,0)=g(t-\tau)$$
entonces la ecuación (2.4) se reduce a
$$ y(t)=\int_{0}^t g(t-\tau) u(\tau) d\tau =\int_{0}^t g(\tau) u(t-\tau) d\tau $$
la cual se conoce como la integral de convolución.

Los siguientes videos ilustran la integral de convolución:

Una de las formas para resolver la integral de convolución consiste en usar la transformada de Laplace, la cual convierte la integral de la convolución en una multiplicación, siempre y cuando existan $\mathcal{L} \left(  g(t)\right)$ y $ \mathcal{L} \left( u(t) \right)$.
$$ \mathcal{L} \left( \int_{0}^t g(\tau) u(t-\tau) d\tau \right)=\mathcal{L} \left(  g(t)\right) \mathcal{L} \left( u(t) \right)$$
En los casos que que se pueda calcular, se obtiene
$$  \int_{0}^t g(\tau) u(t-\tau) d\tau =\mathcal{L}^{-1} \left(\mathcal{L} \left(  g(t)\right) \mathcal{L} \left( u(t) \right)\right)$$

| Respuesta al impulso | Función de transferencia |
| - | - |   
| $g(t)$ | $\mathcal{L}(g(t))$ |
| $\mathcal{L}^{-1}(G(s))$ | $G(s)$ |

Ejemplo [Chen, 2.4]:
Considere un sistema que retarda la unidad de tiempo. Entonces su respuesta al impulso $\delta(t)$ es $\delta(t-1)$. Por lo tanto, la función de transferencia es  
$$ G(s)=\mathcal{L}(\delta(t-1))=\int_0^\infty \delta(t-1) e ^{-st} dt= e^{-st}\Big|_{t=1}=e^{-s}$$
